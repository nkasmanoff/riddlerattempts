From Mahalingam Vaidhyanathan, a probabilistic pub game:

You are throwing darts at a dartboard that has a radius of 1 foot. Due to a gift of miraculous marksmanship, your darts always fall on the board and never outside. (Darts can land on the very edge of the circular board, and if they do they’re considered as landing inside the scoring area.) Furthermore, your chances of hitting any area on the board are exactly proportional to the area of the patch — your darts land according to a uniform probability distribution.

You keep throwing darts until your nth dart hits a location that is less than 1 foot from some other dart. You are then “out,” and n-1 is your final score. Here are three questions of increasing difficulty about this game:

What is the maximum possible score?
What is the probability of getting a score greater than 1 (i.e., that the second dart falls more than 1 foot away from the first)?
What is the expected value of your score?

Intuitively, it seems like the max number of darts that could possibly within this board while all being at least 1 foot apart is 7. There should be one on the border of the board, each separated by 60 degrees, along with one in the center. In this code I'm hoping to prove this or find a different answer. 

The procedure I'm going to use is by randomly generating N darts that all land on the board, and test each one, and see the max number of other darts that also fit on the board. This is going to use some nested for loops where each time a dart is tested, it runs through all the other possible darts, sees how many possibly work, and so on and so on. 

```python
%matplotlib notebook
import numpy as np
import random as rand
import matplotlib.pyplot as plt
```

```python
def dist(r1,r2):
    #find the distance between two points
    return np.sqrt((r2[0] - r1[0])**2 + (r2[1] - r1[1])**2)
```

```python
#generate all the darts thrown, has an arbitrary number Ndarts. Should fit within a circle of radius 1 foot. 

Ndarts = 500
xdart = []
ydart = []
for i in range(Ndarts):
    xrand = rand.uniform(-1,1)
    yrand = rand.uniform(-1,1)
    if (xrand**2 + yrand**2)<=1:
        xdart.append(xrand)
        ydart.append(yrand)
#forgot to limit this to be within a circle!
print(len(xdart))
```
