```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1.0, 1.0, 20)
y = np.linspace(-1.0, 1.0, 20)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2
plt.contour(X,Y,F,[0],'b')
```

```python

x = np.linspace(-1, 1, 11)
y = []
yneg = []
for i in range(len(x)):
    y.append(np.sqrt(1-x[i]**2))
    yneg.append(-np.sqrt(1-x[i]**2))
plt.plot(x, y,'o')
plt.plot(x, yneg,'o')
```
